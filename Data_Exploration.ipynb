##Data Flaws/Weaknesses and Cleaning



**Flaws/Weaknesses**

There are missing values in the data that we found. To figure out if there were missing fields, we first read the csv file into a dataframe, then iterating over each row we counted which ones had missing values. Here is an example below:

```python
import pandas as pd
df = pd.read_csv('movie_metadata.csv')


count = 0 # this is the counter initialized to zero
for index, row in df.iterrows(): # iterate for each row in the dataframe
    if pd.isnull(row).any(): # if there are any missing values in the row count them
        count += 1

print 'Number of rows with missing values:',count
```



Another part we need to do is create a new column, Profit. We have the movie's bduget and the movie's gross, but we want to see how much money a movie actually made or profited. So we will calculate the difference of these two columns to create the Profit column. 

```python
#here we will create new column Profit
import pandas as pd
import numpy as np
df['profit'] = df.gross - df.budget

# print df.describe()
df = df[np.isfinite(df['profit'])]
df = df[np.isfinite(df['actor_3_facebook_likes'])]

df.describe()
```

```python
#Begin to concat the data and eliminate the outliers
#create a new dataframe with only the columns we need to explore
new_df = pd.concat([df['profit'], df['gross'], df['budget'], df['cast_total_facebook_likes'], df['imdb_score']], axis=1, keys=['profit', 'gross', 'budget','cast_total_facebook_likes', 'imdb_score'])


#cut out outliers of the profit column to only have the min and max be 
# positive or negative $500,000,000
new_df = new_df[np.abs(new_df['profit']) < 100000000]

#cut out outliers of facebook likes
new_df = new_df[new_df['cast_total_facebook_likes'] < 100000 ]

#show the statistic summary of the data
new_df.describe()
```

However, zooming in on the data, it is interesting to find that when the Facebook likes are below are certain threshold, there is a correlation between the amount of likes and the movie's profit. When examining Facebook likes above 60,000, the correlation becomes nonexistant.
